```python
df = pd.read_csv("sales_data_sample.csv", encoding='latin1')
```

```python
print("First 5 rows:")
display(df.head())
```

```python
print("\nDataset Info:")
df.info()
```

```python
print("\nSummary Statistics:")
display(df.describe())
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
print("\nTotal Sales by Product:")
display(df.groupby("PRODUCTLINE")["SALES"].sum())
```

```python
print("\nTotal Sales by Country:")
display(df.groupby("COUNTRY")["SALES"].sum())
```

```python
print("\nSummary Statistics:")
display(df.describe())

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt

product_sales = df.groupby("PRODUCTLINE")["SALES"].sum()

plt.figure()
product_sales.plot(kind='bar')
plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
```

```python
region_sales = df.groupby("COUNTRY")["SALES"].sum()

import matplotlib.pyplot as plt
plt.figure()
region_sales.plot(kind='bar')
plt.title("Total Sales by Country")
plt.xlabel("Country")
plt.ylabel("Sales")
plt.show()
```

```python
df["ORDERDATE"] = pd.to_datetime(df["ORDERDATE"])
sales_trend = df.groupby("ORDERDATE")["SALES"].sum()

plt.figure()
sales_trend.plot(kind='line')
plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
```
